```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(df[['Age']],df.InsuranceBought,test_size=0.1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model.fit(X_train,Y_train)
```
